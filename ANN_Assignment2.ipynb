```python
#Importing the Keras library
from keras.models import Sequential
from keras.layers import Dense
```

```python
seed =7
np.random.seed(seed)
```

```python
#Splitting into X & Y variables
x = df.iloc[ : , :-1].values
y = df.iloc[:, 1].values
```

```python
#Creating the appropriate model with the appropriate number of layers
model = Sequential()
model.add(Dense(16, input_dim=13,  activation='relu')) #1st layer
model.add(Dense(12,  activation='relu')) #2nd layer
model.add(Dense(8, activation='relu')) #3rd layer
model.add(Dense(6, activation='relu')) #4th layer
model.add(Dense(3, activation='relu')) #5th layer
model.add(Dense(1, activation='sigmoid')) #6th layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, validation_split=0.30, epochs=110, batch_size=10)
```

```python
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
history.history.keys()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarizing history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
