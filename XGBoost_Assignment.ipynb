```python
# Converting the Species column into integers using the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'Species' column to numeric format
df['Species'] = label_encoder.fit_transform(df['Species'])

df
```

```python
target = df['Species']
df.drop(columns={'Id','Species'}, inplace=True)
```

```python
df
```

```python
#Giving the parameters for the RandomForest model
param_dist = {
    'n_estimators': randint(100, 1000),
    'max_features': ['auto', 'sqrt'],
    'max_depth': randint(5, 50),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
    'bootstrap': [True, False]
}
```

```python
# Fitting the RandomForest model
model_dt = RandomForestClassifier()

# Performing RandomizedSearchCV
clf = RandomizedSearchCV(model_dt, param_distributions=param_dist, n_iter=100, cv=7, random_state=45, n_jobs=-5)
clf.fit(df, target)
```

```python
print("Best hyperparameters found by RandomizedSearchCV:")
print(clf.best_params_)
```

```python
# spliting the  data into X and y
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
# to split the dataset intp training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#Printing to check the shape of the spilt datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
y_train = y_train.astype(int)
```

```python
#Fitting the model (the training batch) into the XGBoostClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
#Predicting the testing batch/split
y_pred = model.predict(X_test)
```

```python
#Converting the y_test into intergers as y_pred is a multi class classifier
y_test = y_test.astype(int)
```

```python
#Checking the accuracy of the trained model to check how accurately it can predict the outcome
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy of the XGBppst Model: ",accuracy)
```
